```python
import os
from googleapiclient.discovery import build

# Set your YouTube Data API key here
API_KEY = "AIzaSyA0vaLuQ-hogdwCoGY5-XcP8itCY4MwPzQ"

# Initialize the YouTube Data API client
youtube = build('youtube', 'v3', developerKey=API_KEY)

def search_channels_by_keywords(keywords, max_results=10):
    # Search for channels using the provided keywords
    search_response = youtube.search().list(
        q=keywords,
        type='channel',
        part='id,snippet',
        maxResults=max_results
    ).execute()

    # Extract channel information from the search response
    channels = []
    for item in search_response.get('items', []):
        channel = {
            'channel_id': item['id']['channelId'],
            'title': item['snippet']['title'],
            'description': item['snippet']['description']
        }
        channels.append(channel)

    return channels

if __name__ == "__main__":
    search_keywords = "IIIT NUZVID"
    max_results = 100

    found_channels = search_channels_by_keywords(search_keywords, max_results)

    print(f"Found {len(found_channels)} channels related to '{search_keywords}':")
    for channel in found_channels:
        print(f"Title: {channel['title']}")
        print(f"Channel ID: {channel['channel_id']}")
        print(f"Description: {channel['description']}")
        print("=" * 30)
```
